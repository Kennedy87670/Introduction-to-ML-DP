```python
# import matplotlib  and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1.  preparing data
x = [1,2,3,4]
y = [ 11, 22, 33, 44]

#2 setup plot
fig, ax= plt.subplots(figsize=(10,10)) #figzise means the region num in x, y weight and height

#3. plot data
ax.plot(x, y)

#4 customize plot
ax.set(title="simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5 save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
#which ne should i use (pyplot vs matplotlib oo method)
*when plotting something quickly its okay to use the pyplot method
*when plotting something more advanced , use the oo method
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#pyplot method mixed with pyplot
over_50.plot(kind= "scatter",
            x= "age",
            y= "chol",
            c= "target");
```

```python
#OO method
fix, ax =plt.subplots(figsize= (10, 6))
over_50.plot(kind="scatter",
            x="age",
            y= "chol",
            c= "target",
            ax=ax);

#inrease the liit to the x axis
ax.set_xlim((45, 100));
```

```python
#OO Mthod from scratch
fig,ax = plt.subplots(figsize = (10, 6))

#plot the data
scatter= ax.scatter(x= over_50["age"],
                   y=over_50["chol"],
                   c=over_50['target']);
#customise the plot
ax.set(title= "Heart Disease and Cholesterol Levels",
      xlabel=" Age",
      ylabel= "cholesterol");
#add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle= "--");
```

```python
over_50.head()
```

```python
#subplots of chol, age, thalach
fig, (ax0, ax1)= plt.subplots(nrows= 2,
                             ncols=1,
                             figsize= (10,10),
                              sharex=True)

#add data at ax0
scatter= ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

#customise ax0
ax0.set(title="Heart Disease and Cholesterol Levls",
       ylabel="cholesterol")

#add legend for ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle= "--");

#add to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])
#customize ax1
ax1.set(title="Heart Disease and Max heart rate",
       xlabel="Age",
       ylabel="Max heart rate");

#add legend for ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

#add figure to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
#set the style
plt.style.use('seaborn-whitegrid')

#OO Mthod from scratch
fig,ax = plt.subplots(figsize = (10, 6))

#plot the data
scatter= ax.scatter(x= over_50["age"],
                   y=over_50["chol"],
                   c=over_50['target'],
                   cmap='winter');#this changes our colour scheme
#customise the plot
ax.set(title= "Heart Disease and Cholesterol Levels",
      xlabel=" Age",
      ylabel= "cholesterol");
#add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle= "--");
```

```python
#customizing the y and x limitation


#subplots of chol, age, thalach
fig, (ax0, ax1)= plt.subplots(nrows= 2,
                             ncols=1,
                             figsize= (10,10),
                              sharex=True)

#add data at ax0
scatter= ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap='winter')

#customise ax0
ax0.set(title="Heart Disease and Cholesterol Levls",
       ylabel="cholesterol")

#change the x axis limit
ax0.set_xlim([50,80])

#add legend for ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle= "--");

#add to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap='winter')
#customize ax1
ax1.set(title="Heart Disease and Max heart rate",
       xlabel="Age",
       ylabel="Max heart rate");

#change ax1 x axis limits
ax1.set_xlim([50,80])
#ax1.set_xlim([60,200])

#add legend for ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

#add figure to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
